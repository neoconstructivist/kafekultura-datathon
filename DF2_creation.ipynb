```python
import pandas as pd
import numpy as np
from dateutil.parser import parse
import datetime
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
d1 = pd.read_csv('data/first_df.csv')
```

```python
df2 = pd.DataFrame(columns=['D', 'Weekday','Avg_duration', 'Avg_cash', 'Num_ppl'])
```

```python
items_str = []
for i in range(1, 34):
    items_str.append('item_' + str(i))
df_cols = ['D', 'T_period', 'Weekday','Avg_duration', 'Avg_cash', 'Num_ppl', 'Total_sum']
df_cols.extend(items_str)
```

```python
while (d1.shape[0]>0):
    temp_series = pd.Series(index=df_cols)
    dt=  d1.loc[0, 'Date']
    tp = d1.loc[0,'T_period']
    
    temp_series['D'] = d1.loc[0, 'Date']
    temp_series['T_period'] = d1.loc[0, 'T_period']
    temp_series['Weekday'] = d1.loc[0, 'Weekday']
    temp_series['Avg_duration'] = np.round(d1.Duration[((d1.Date==dt) & (d1.T_period==tp))].mean(), decimals=1)
    temp_series['Avg_cash'] = np.round(d1.Total_sum[((d1.Date==dt) & (d1.T_period==tp))].mean(), decimals=1)
    temp_series['Num_ppl'] = d1.Num_guests[((d1.Date==dt) & (d1.T_period==tp))].sum()
    temp_series['Total_sum'] = d1.Total_sum[((d1.Date==dt) & (d1.T_period==tp))].sum()
    
    for i in range(1, 34):
        ind = 'item_' + str(i)
        temp_series[ind] = d1[ind][((d1.Date==dt) & (d1.T_period==tp))].sum()
    
    
    df2 = df2.append(temp_series, ignore_index=True)
    d1 = d1.drop(d1[((d1.Date==dt) & (d1.T_period==tp))].index, axis=0)
    d1 = d1.reset_index(drop=True)
```

```python
for i in range(df2.shape[0]):
    df2.D[i] = parse(df2.D[i]).date()
```
